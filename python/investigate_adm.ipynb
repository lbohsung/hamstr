```python
import matplotlib.pyplot as plt
import numpy as np
import arviz as az
from posterior_methods import get_posterior_ages
```

```python
def plot_age_models(idata):
    posterior_ages_df = get_posterior_ages(idata)
    depths = np.array(posterior_ages_df[posterior_ages_df.iter == 1].depth)
    ages = np.array(posterior_ages_df.age).reshape(1000, -1)
    ages_mean = np.array(np.mean(ages, axis=0))
    plt.plot(depths, ages_mean, c="C3", zorder=3)
    plt.plot(depths, ages[::10].T, c="black", alpha=0.1)
    plt.show()

idata = az.from_netcdf("model_results.nc")
plot_age_models(idata)
```
